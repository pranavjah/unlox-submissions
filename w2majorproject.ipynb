### Q1. Star Pattern Write a Python program to print a square star pattern of size n × n using a loop. Example (n = 4):

```python
# Taking input from user
n = int(input("Enter the value of n: "))

# Printing square star pattern
for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()   # move to next line
```

### Q2. Reverse a String Write a function that takes a string as input and returns the reversed string without using built-in reverse functions.

```python
# Function to reverse a string manually
def reverse_string(text):
    reversed_text = ""

    # Loop from last character to first
    for i in range(len(text) - 1, -1, -1):
        reversed_text += text[i]

    return reversed_text


# Taking input
user_input = input("Enter a string: ")
result = reverse_string(user_input)

print("Reversed string is:", result)
```

### Q3. Function and Logic Create a user-defined function that:

- Accepts a number as input
- Returns whether the number is a Palindrome number or not

```python
# Function to check palindrome number
def is_palindrome(number):
    original = str(number)
    reversed_num = ""

    # Reverse manually
    for i in range(len(original) - 1, -1, -1):
        reversed_num += original[i]

    if original == reversed_num:
        return "Palindrome Number"
    else:
        return "Not a Palindrome Number"


# Taking input
num = int(input("Enter a number: "))
print(is_palindrome(num))
```

## Section B – NumPy

### Q4. NumPy Array Operations
- Create a NumPy array of integers from 1 to 20
- Reshape it into a 4 × 5 matrix
- Find:
  - Mean of each column
  - Maximum value of each row

```python
import numpy as np

# Create array from 1 to 20
arr = np.arange(1, 21)

# Reshape into 4x5 matrix
matrix = arr.reshape(4, 5)

print("4x5 Matrix:")
print(matrix)

# Mean of each column
column_mean = np.mean(matrix, axis=0)
print("\nMean of each column:")
print(column_mean)

# Maximum value of each row
row_max = np.max(matrix, axis=1)
print("\nMaximum value of each row:")
print(row_max)
```

### Q5. Conditional Operations Using NumPy:
- Replace all values greater than 10 with -1
- Display the modified array

```python
# Replace all values greater than 10 with -1
modified_matrix = matrix.copy()

modified_matrix[modified_matrix > 10] = -1

print("Modified Matrix:")
print(modified_matrix)
```

### Q6. Data Loading and Inspection
- Load the dataset using Pandas
- Display:

```python
import pandas as pd

# Load dataset
df = pd.read_csv("4892262-flipkart_com-ecommerce_sample.csv")

# First 5 rows
print("First 5 rows:")
display(df.head())

# Dataset shape
print("Dataset Shape:", df.shape)

# Column names
print("\nColumn Names:")
print(df.columns)

# Basic information
print("\nDataset Info:")
df.info()
```

### Q7. Data Cleaning Perform the following:
- Handle missing values appropriately
- Convert price and rating columns to numeric types
- Remove duplicate records (if any)

```python
# Checking missing values
print("Missing values before cleaning:")
print(df.isnull().sum())

# Handling missing values
df = df.dropna(subset=['retail_price', 'discounted_price', 'product_rating'])

# Convert price and rating columns to numeric
df['retail_price'] = pd.to_numeric(df['retail_price'], errors='coerce')
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['product_rating'] = pd.to_numeric(df['product_rating'], errors='coerce')

# Remove duplicates
df = df.drop_duplicates()

print("\nMissing values after cleaning:")
print(df.isnull().sum())

print("\nShape after cleaning:", df.shape)
```

### Q8. Data Analysis Answer the following using Pandas:
- Find the top 5 product categories by average rating
- Identify the 5 most expensive products

```python
# Top 5 product categories by average rating
top_categories = (
    df.groupby('product_category_tree')['product_rating']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

print("Top 5 Categories by Average Rating:")
print(top_categories)


# 5 Most expensive products (based on retail price)
most_expensive = df.sort_values(by='retail_price', ascending=False).head(5)

print("\n5 Most Expensive Products:")
display(most_expensive[['product_name', 'retail_price']])
```

## Section D – Exploratory Data Analysis

### Q9. Visualization Create the following plots:
- Distribution of product prices
- Count plot of product categories  

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of product prices
plt.figure()
plt.hist(df['retail_price'], bins=30)
plt.title("Distribution of Product Prices")
plt.xlabel("Retail Price")
plt.ylabel("Frequency")
plt.show()


# Count plot of product categories (Top 10 for clarity)
top_10_categories = df['product_category_tree'].value_counts().head(10)

plt.figure()
sns.barplot(x=top_10_categories.values, y=top_10_categories.index)
plt.title("Top 10 Product Categories")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```
